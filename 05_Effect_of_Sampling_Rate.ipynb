Consider the analog signal: xa(t)=3cos(200πt)+5sin(600πt)+10cos(1200πt). Show the effect of sampling rate.

```python
import numpy as np
import matplotlib.pyplot as plt

A = [3, 5, 10]
f = [100, 300, 600]

t_analog = np.linspace(0, 0.05, 1000)

def sig_func(A1, A2, A3, f1, f2, f3, t_analog):
    return A1 * np.cos(2 * np.pi * f1 * t_analog) + A2 * np.sin(2 * np.pi * f2 * t_analog) + 10 * np.cos(2 * np.pi * f3 * t_analog)

signal = sig_func(A[0], A[1], A[2], f[0], f[1], f[2], t_analog)

plt.figure(figsize=(10, 6))
plt.subplot(4, 1, 1)
plt.plot(t_analog, signal)
plt.title('Original Signal')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

fs = 2000
Ts = 1 / fs
t = np.arange(0, 0.05, Ts)
signal = sig_func(A[0], A[1], A[2], f[0], f[1], f[2], t)
plt.subplot(4, 1, 2)
plt.stem(t, signal)
plt.plot(t, signal)
plt.title('Sampling at 2000')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

fs = 600
Ts = 1 / fs
t = np.arange(0, 0.05, Ts)
signal = sig_func(A[0], A[1], A[2], f[0], f[1], f[2], t)
plt.subplot(4, 1, 3)
plt.stem(t, signal)
plt.plot(t, signal)
plt.title('Sampling at 600')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
