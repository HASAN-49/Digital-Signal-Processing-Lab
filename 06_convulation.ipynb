The impulse response of a discrete-time LTI system is h(n)={u(n)−u(n−5)}. Determine the output of the system for the input x[n]=u(n), using the convolution sum.

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(n):
    return np.where(n >= 0, 1, 0)

n = np.arange(0, 21)
x = unit_step(n)
h = unit_step(n) - unit_step(n - 5)

def convulation(x, h):
    lnx = len(x)
    lnh = len(h)
    lny = lnx + lnh - 1
    y = np.zeros(lny)

    for n in range(lny):
        for k in range(lnx):
            if 0 <= n - k < lnh:
                y[n] += x[k] * h[n - k]
    return y

y = convulation(x, h)
n_y = np.arange(len(y))

plt.figure(figsize=(10, 6))
plt.stem(n_y, y)
plt.title('title')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```
