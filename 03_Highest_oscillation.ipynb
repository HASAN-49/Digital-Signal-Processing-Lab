Show that the highest rate of oscillation in a discrete-time sinusoidal is obtained when ω=π.

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 11, 1)
omega1 = np.pi
omega2 = np.pi / 4

wave1 = np.cos(n * omega1)
wave2 = np.cos(n * omega2)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.stem(n, wave1, label='omega=pi')
plt.title('Cosine Wave')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(n, wave2, label='omega=pi/4')
plt.title('Cose wave')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
