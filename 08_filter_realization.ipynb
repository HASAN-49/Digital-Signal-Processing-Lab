Filter realization using 6-point averaging, 6-point differencing equations.

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(200)
x = np.sin(0.1 * np.pi * n) + 0.5 * np.random.randn(200)

def averaging(signal):
    kernel = np.ones(6) / 6
    return np.convolve(signal, kernel, mode='same')

def differ(signal):
    kernel = np.array([1, -1, 1, -1, 1, -1]) / 6
    return np.convolve(signal, kernel, mode='same')

y_avg = averaging(x)
y_dif = differ(x)

plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(n, x, label='Original Signal')
plt.title('Original Signal')
plt.grid()
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(n, y_avg, color='red', label='6-point averaging')
plt.title('6 point averaging')
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(n, y_dif, color='green', label='6-point dif')
plt.title('6 poing differencing')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
