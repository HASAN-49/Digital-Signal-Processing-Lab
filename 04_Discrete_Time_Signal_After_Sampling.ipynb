Consider the continuous-time analog signal x(t)=3cos(100πt). Sample the analog signal at 200 Hz and 75 Hz. Show the discrete-time signal after sampling. ⟹ realization

```python
import numpy as np
import matplotlib.pyplot as plt

A = 3
f = 50
omega = 2 * np.pi * f

n = np.arange(0, 41, 1)
fs1 = 200
Ts1 = 1 / fs1
t = n * Ts1
x1 = A * np.cos(2 * np.pi * f * t)
t_analog = np.linspace(0, 40 * Ts1, 1000)
y_analog = A * np.cos(2 * np.pi * f * t_analog)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t_analog / Ts1, y_analog, 'k--')
plt.stem(n, x1)
plt.title('fs1=200')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

n = np.arange(0, 16, 1)
fs2 = 75
Ts2 = 1 / fs2
t = n * Ts2
x2 = A * np.cos(2 * np.pi * f * t)
t_analog = np.linspace(0, 15 * Ts2, 1000)
y_analog = A * np.cos(2 * np.pi * f * t_analog)

plt.subplot(2, 1, 2)
plt.plot(t_analog / Ts2, y_analog, 'k--')
plt.stem(n, x2)
plt.title('fs2=75')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
