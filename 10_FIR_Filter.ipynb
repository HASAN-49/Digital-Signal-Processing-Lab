Design a low pass FIR filter to remove high-frequency noise from a signal using convolution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, lfilter

fs = 1000
Ts = 1 / fs
t = np.arange(0, 1, Ts)
signal = np.sin(2 * np.pi * 5 * t)
noise = 0.5 * np.sin(2 * np.pi * 100 * t) + 0.3 * np.random.randn(len(t))
signal_noisy = signal + noise

curoff = 20
num_taps = 101
fir_coeff = firwin(num_taps, curoff, fs=fs)
filtered_signal = lfilter(fir_coeff, 1.0, signal_noisy)

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal_noisy)
plt.title('Noisy Signal')

plt.subplot(3, 1, 2)
plt.plot(t, filtered_signal, color='green', label='filtered_signal')
plt.plot(t, signal, 'k--', label='Original Signal')
plt.title('Filtered Signal')
plt.legend()

# plt.subplot(3, 1, 3)

plt.tight_layout()
plt.show()
```
