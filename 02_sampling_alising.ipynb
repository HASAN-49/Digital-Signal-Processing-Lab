Demonstrates the effect of sampling, aliasing.

```python
import numpy as np
import matplotlib.pyplot as plt

f = 10
t = np.linspace(0, 1, 1000)
y = np.sin(2 * np.pi * f * t)

plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, y)
plt.title('Original signal')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
# plt.legend()

fs_high = 50
fs_low = 15

t_high = np.arange(0, 1, 1 / fs_high)
t_low = np.arange(0, 1, 1 / fs_low)

y_high = np.sin(2 * np.pi * f * t_high)
y_low = np.sin(2 * np.pi * f * t_low)

plt.subplot(3, 1, 2)
plt.plot(t, y, 'k--', label = 'Original Signal')
plt.stem(t_high, y_high, label='f=50')
plt.title('No alising')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, y, 'k--', label = 'Original Signal')
plt.stem(t_low, y_low, label='f=15')
plt.title('Alising')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
