Given x(n)=[1,3,−2,4] y(n)=[2,3,−1,3] z(n)=[2,−1,4,−2] Find the correlation between x(n) & y(n) and y(n) & z(n). ⟹ observe the realization.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 3, -2, 4])
y = np.array([2, 3, -1, 3])
z = [2, -1, 4, -2]

def cross_correlation(x, y):
    lnx = len(x)
    lny = len(y)
    corr_len = lnx + lny - 1
    result = []

    for i in range(-lnx + 1, lny):
        sum = 0
        for j in range(lnx):
            k = j - i
            if 0 <= k < lny:
                sum += x[j] * y[k]
        result.append(sum)
    return result

r_xy = cross_correlation(x, y)
r_yz = cross_correlation(y, z)
n = np.arange(-len(x) + 1, len(y))
# print(r_xy)

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.stem(n, r_xy)
plt.title('Cross Correlation')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(n, r_yz)
plt.title('Cross Correlation')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.grid(True)

plt.tight_layout()
plt.show()
```
